```python
import pandas as pd
import numpy as np
```

```python
#numpy basic
arr = np.array([1,2,3,4,5])
print("Numpy array:", arr)
print("Mean:", arr.mean())
print("Sum:", arr.sum())
```

```python
#pandas basic
#creating a dataframe
data = {
    "name": ["John", "Sam", "Adam", "Josh", "Mike"],
    "age": [23, 45, 56, 12, 32],
    "sale": [100, 400, 800, 200, 1000]
}
df = pd.DataFrame(data)
# print("\nDataframe:")
# print(df)

print("\nAge:", df["age"])

print("\nAverage Slaes:", df["sale"].mean())
```

```python
#First Visualization
df.set_index("name")["sale"].plot(kind = "bar", title = "Sale by person")
```

```python
#Similar but easy and qucik
df.plot(x= "name", y = "sale", kind = "bar", title ="Sale by person")
```
